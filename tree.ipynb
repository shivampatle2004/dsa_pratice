```python
def dfs(graph,start,visited=None,order=None):
  if visited is None:
    visited =set()
  if order is None:
    order = []
  visited.add(start)
  order.append(start)
  for neighbour in graph.get(start,[]):
    if neighbour not in visited:
      dfs(graph,neighbour,visited,order)
  return order
```

```python
if __name__ == "__main__":
  graph={
      'A' : ['B','C'],
      'B' : ['D','E'],
      'C' : ['F'],
      'D' : [],
      'E' : ['F'],
      'F' : []
  }
  print(dfs(graph,'A'))
```

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    order = []

    visited.add(start)

    while queue:
        vertex = queue.popleft()
        order.append(vertex)

        for neighbour in graph.get(vertex, []):
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)
    return order
```

```python
new_graph = {
    'a': ['b', 'c'],
    'b': [],
    'c': ['b', 'd', 'g'],
    'd': ['a', 'e', 'h'],
    'e': ['c', 'f'],
    'g': ['h'],
    'h': ['b']
}
```

```python
print(f"BFS traversal for the graph from 'a': {bfs(new_graph, 'a')}")
```
