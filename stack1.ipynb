```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return "Stack is empty"

    def size(self):
        return len(self.items)

# Demonstration of the Stack
my_stack = Stack()
print(f"Is stack empty? {my_stack.is_empty()}")

my_stack.push("{")
my_stack.push("[")
my_stack.push("(")
my_stack.push(")")
my_stack.push("]")
my_stack.push("}")


print(f"Stack after pushes: {my_stack.items}")
print(f"Top element (peek): {my_stack.peek()}")
print(f"Popped element: {my_stack.pop()}")
print(f"Stack after one pop: {my_stack.items}")
print(f"Popped element: {my_stack.pop()}")
print(f"Is stack empty? {my_stack.is_empty()}")
print(f"Popped element: {my_stack.pop()}")
print(f"Stack after all pops: {my_stack.items}")
print(f"Trying to pop from empty stack: {my_stack.pop()}")
```

### Explanation of the Stack Code:

*   **`__init__(self)`**: Initializes an empty list (`self.items`) to store the stack elements.
*   **`is_empty(self)`**: Returns `True` if the stack contains no elements, `False` otherwise.
*   **`push(self, item)`**: Adds an `item` to the end of the `items` list, effectively placing it on top of the stack.
*   **`pop(self)`**: Removes and returns the last element from the `items` list (the top of the stack). It includes a check to prevent popping from an empty stack.
*   **`peek(self)`**: Returns the last element of the `items` list (the top of the stack) without removing it. Also includes an empty stack check.
*   **`size(self)`**: Returns the number of elements currently in the stack.

```python
# Check if a user-defined input is balanced
expression_to_check = input("Enter an expression with brackets (e.g., '{[()]}'): ")

# Ensure the Stack class is defined (it was in a previous cell)
# If it wasn't, we would need to redefine it here or ensure it's run first.

matching_pairs = {
    ')': '(',
    ']': '[',
    '}': '{'
}

balance_stack = Stack()
print(f"Checking expression: {expression_to_check}")
is_balanced = True

for char in expression_to_check:
    if char in ['{', '[', '(']:
        balance_stack.push(char)
        print(f"Pushed '{char}'. Current stack: {balance_stack.items}")
    elif char in [')', ']', '}']:
        if balance_stack.is_empty():
            is_balanced = False
            print(f"Mismatch: Closing '{char}' found, but stack is empty.")
            break

        top_element = balance_stack.pop()
        print(f"Popped '{top_element}'. Current stack: {balance_stack.items}")
        if matching_pairs[char] != top_element:
            is_balanced = False
            print(f"Mismatch: Expected '{matching_pairs[char]}' but got '{top_element}' for closing '{char}'.")
            break

if not balance_stack.is_empty():
    is_balanced = False
    print(f"Remaining unclosed brackets in stack: {balance_stack.items}")

if is_balanced:
    print(f"\nThe expression '{expression_to_check}' is BALANCED.")
else:
    print(f"\nThe expression '{expression_to_check}' is NOT BALANCED.")
```

```python
#Infix to Postfix
def precedence(symbol):
  if symbol == '^':
    return 3
  elif symbol in['*','/']:
    return 2
  elif symbol in['+','-']:
    return 1
  else:
    return 0

def infix_postfix(infix):
  stack=[]
  postfix=""
  stack.append('(')
  infix += ')'

  for symbol in infix:
    if symbol == ' ':
      continue
    if symbol == '(':
      stack.append(symbol)
    elif symbol.isalnum():
      postfix +=symbol
    elif symbol==')':
      while stack and stack[-1]!= '(': # Ensure stack is not empty before peeking
        postfix += stack.pop()
      stack.pop() # Pop the '('
    else: # Operator
      while stack and precedence(stack[-1]) >= precedence(symbol):
        postfix += stack.pop()
      stack.append(symbol)
  return postfix

# Demonstration
infix_expression ="A+B*C"
print("Infix Expression :",infix_expression)
postfix_expression=infix_postfix(infix_expression)
print("Postfix Expression : ",postfix_expression)
```
